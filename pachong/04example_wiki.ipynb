## 合并整理

```python

def get_html(url):
    response = requests.get(url, headers=header)
    bssoup = BeautifulSoup(response.text, 'html.parser')
    return bssoup


def scrapy_1html(url, ls_cnt, ls_title, ls_a, ls_p, ls_time, ls_publisher, ls_rate, ls_quote):
    soup = get_html(url)
    # cnt = 0
    for item in soup.findAll(attrs={'class': 'item'}):
        # number
        # cnt += 1
        # title
        title = item.find(attrs={'class': 'pl2'}).text.strip().replace("\n", "").replace(" ", "")  # title
        # Four
        four = item.find(attrs={'class': 'pl'}).text.split('/')
        # author
        author = four[0]
        # price
        price = four[-1]
        # time
        time = four[-2]
        # publisher
        publisher = four[-3]
        # rating_nums
        rating_nums = item.find(attrs={'class': 'rating_nums'}).text
        # quote
        try:
            quote = item.find(attrs={'class': 'inq'}).text
        except:
            quote = ''

        # ls_cnt.append(cnt)
        ls_title.append(title)
        ls_a.append(author)
        ls_p.append(price)
        ls_time.append(time)
        ls_publisher.append(publisher)
        ls_rate.append(rating_nums)
        ls_quote.append(quote)
    return ls_cnt, ls_title, ls_a, ls_p, ls_time, ls_publisher, ls_rate, ls_quote
```

```python
ls_cnt = []
ls_title = []
ls_a = []
ls_p = []
ls_time = []
ls_publisher = []
ls_rate = []
ls_quote = []

for page in range(0, 250, 25):
    newurl = douban + str(page)
    scrapy_1html(newurl, ls_cnt, ls_title, ls_a, ls_p, ls_time, ls_publisher, ls_rate, ls_quote)
# print(dic)
```

```python
# 构建字典

dic = {'title': ls_title, 'author': ls_a, 'price': ls_p, 'time': ls_time,
       'publisher': ls_publisher, 'rate': ls_rate, 'quote': ls_quote}
df = pd.DataFrame(dic)
print(df)
```

```python
random.seed(datetime.datetime.now())


def getLinks(articleUrl):
    html = urlopen('http://en.wikipedia.org{}'.format(articleUrl))
    bs = BeautifulSoup(html, 'html.parser')
    return bs.find('div', {'id': 'bodyContent'}).find_all('a', href=re.compile('^(/wiki/)((?!:).)*$'))


links = getLinks('/wiki/Kevin_Bacon')
while len(links) > 0:
    newArticle = links[random.randint(0, len(links) - 1)].attrs['href']
    print(newArticle)
    links = getLinks(newArticle)
```
