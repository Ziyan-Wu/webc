```python
import requests
```

```python
url = 'https://baidu.com'
# https://ssr1.scrape.center/page/4

headers = {
    'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/48.0.2564.116 Safari/537.36'
}
```

在使用python爬虫爬取数据的时候，经常会遇到一些网站的反爬虫措施，一般就是针对于headers中的User-Agent，
如果没有对headers进行设置，User-Agent会声明自己是python脚本,而如果网站有反爬虫的想法的话，必然会拒绝这样的连接。
而修改headers可以将自己的爬虫脚本伪装成浏览器的正常访问，来避免这一问题。

```python
response = requests.get(url, headers)
response.encoding = 'utf-8'
# 以字符串形式返回网页源码!!!
print(response.text)
```

```python
# 返回二进制的数据
print(response.content)
```

```python
print(response.status_code)
```

```python
print(response.headers)
```

```python
print(response.url)
```

```python
print(response.read())
```
